```python
import pandas as pd
import json
```

```python
coursesfile = open("coursesData.json")
coursesJson = json.loads(coursesfile.read())
coursesfile.close()

coursesMin = []
sectionsObj = []

for course in coursesJson:
	cid = course['course_id']

	sections_ids = ["TBA"] if course['sections']== [] else [ s['section_id'] for s in course['sections']]

	for s in course['sections']:
		meet0 = s["meetings"]
		meet =[]

		for m in meet0: 
			meet.append([ m['classtype'], m['days'], m['start_time'], m['end_time']]) 
		
		sectionsObj.append({
		"section_id": s['section_id'], 
		"meetings": meet ,
		"seats": s['seats'] 
		})	
	
	coreqs = course['relationships']['coreqs']
	prereqs = course['relationships']['prereqs']

	coursesMin.append({
	"id": str(course["course_id"]), 
	"name": str(course["name"]), 
	"desc": str(course["description"]),
	"dept_id": str(course["dept_id"]), 
	"credits": str(course["credits"]),
	"coreqs": str(coreqs),
	"prereqs": str(prereqs), 
	"section_ids": sections_ids,
	})


with open("coursesMin.json", "w") as f:
	json.dump(coursesMin, f)

with open("sections.json", "w") as f:
	json.dump(sectionsObj, f)
```
